```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Datos proporcionados en listas
x1 = [1, 1, 2, 3, 1, 2, 3, 3]
x2 = [0, 1, 1, 2, 2, 3, 3, 1]
y = [3.2, 6, 2.2, 2.5, 6.5, 6.6, 3.5, 0.2]

# Convertir las listas a arreglos NumPy
x1 = np.array(x1)
x2 = np.array(x2)
y = np.array(y)
A = np.vstack([x1, x2, np.ones(len(x1))]).T


coef_x1, coef_x2, intercept = np.linalg.lstsq(A, y, rcond=None)[0]

# Calcular el coeficiente de correlación (r)
corr_matrix = np.corrcoef(np.vstack([x1, x2, y]))


r_x1_y = corr_matrix[0, 2]
r_x2_y = corr_matrix[1, 2]

print("Coeficientes de la función lineal ajustada:")
print(f"coef_x1: {coef_x1}, coef_x2: {coef_x2}, intercept: {intercept}")

print("Coeficientes de correlación:")
print(f"r_x1_y: {r_x1_y}, r_x2_y: {r_x2_y}")


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x1, x2, y, c='r', marker='o')

x1_range = np.arange(min(x1), max(x1) + 1, 0.1)
x2_range = np.arange(min(x2), max(x2) + 1, 0.1)
x1_grid, x2_grid = np.meshgrid(x1_range, x2_range)

y_grid = coef_x1 * x1_grid + coef_x2 * x2_grid + intercept


ax.plot_surface(x1_grid, x2_grid, y_grid, alpha=0.5)



plt.show()
```
